## 2. Some examples
### Example 1. Open 3 `TABs` : `youtube.com`, `github.com` and `facebook.com`
In this example, you are asked to:
- Open 3 given url-links in 3 tabs where `youtube.com` as the `main_window`
- Take the screen-shot after open the last tab : `facebook.com`

```python
## Step 1. determine your url-links
first_tab_link = 'https://www.youtube.com/
second_tab_link = 'https://github.com/'
third_tab_link = 'https://www.facebook.com/'

## Step 2. access & save youtube as the main_window
driver.get(first_tab_link)
main_window = driver.current_window_handle
print(main_window)

## Step 3. open the 2nd tab
driver.execute_script("window.open();")
driver.switch_to_window(driver.window_handles[1])
driver.get(second_tab_link)

## Step 4. open the 3rd tab
driver.execute_script("window.open();")
driver.switch_to_window(driver.window_handles[2])
driver.get(third_tab_link)

## Step 5. screenshot
driver.save_screenshot('screenshot.png')
Image('screenshot.png')
```

### Example 2. Open a new `TAB` of the url of gmail on the `home page of Google`.

```python
## Step 1. access the home-page of Google
first_tab_link = 'https://www.google.com/'
driver.get(first_tab_link)

## Step 2. save the main_window
main_window = driver.current_window_handle

## Step 3. obtain url of gmail on the home page of Google
second_tab_link = driver.find_element_by_xpath('//*[@id="gbw"]/div/div/div[1]/div[1]/a').get_attribute("href")
print(second_tab_link)

## Step 4. Open the new tab of gmail
driver.execute_script("window.open();")
driver.switch_to_window(driver.window_handles[1])
driver.get(second_tab_link)

## Step 5
driver.save_screenshot('screenshot.png')
Image('screenshot.png')
```
